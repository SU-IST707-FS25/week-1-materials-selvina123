## Problem 5

Using the taxi trip data in the `2019_Yellow_Taxi_Trip_Data.csv` file, resample the data to an hourly frequency based on the dropoff time. Calculate the total `trip_distance`, `fare_amount`, `tolls_amount`, and `tip_amount`, then find the 5 hours with the most tips.

```python
taxi_data['tpep_dropoff_datetime'] = pd.to_datetime(taxi_data['tpep_dropoff_datetime'])

hourly = (
    taxi_data
    .set_index('tpep_dropoff_datetime')
    .resample('H')
    .agg({
        'trip_distance': 'sum',
        'fare_amount': 'sum',
        'tolls_amount': 'sum',
        'tip_amount': 'sum'
    })
)

top_5_tips = hourly.sort_values(by='tip_amount', ascending=False).head(5)
top_5_tips
```

## Problem 6

Using the TSA traveler throughput data in the `tsa_melted_holiday_travel.csv` file, create box plots for traveler throughput for each year in the data. Hint: Pass `kind='box'` into the `plot()` method to generate box plots.

```python
import matplotlib.pyplot as plt

# Load TSA data
tsa_data = pd.read_csv("data/tsa_melted_holiday_travel.csv")

# Boxplot by year
tsa_data.plot(kind='box', column='travelers', by='year', figsize=(8,6))
plt.title("TSA Traveler Throughput by Year")
plt.suptitle("")
plt.ylabel("Travelers")
plt.show()
```

## Problem 7
Using the TSA traveler throughput data in the `tsa_melted_holiday_travel.csv` file, create a heatmap that shows the 2019 TSA median traveler throughput by day of week and month.

```python
import seaborn as sns

tsa_2019 = tsa_data[tsa_data['year'] == 2019].copy()
tsa_2019['date'] = pd.to_datetime(tsa_2019['date'])
tsa_2019['day_of_week'] = tsa_2019['date'].dt.day_name()
tsa_2019['month'] = tsa_2019['date'].dt.month_name()


heatmap_data = (
    tsa_2019.groupby(['day_of_week', 'month'])['travelers']
    .median()
    .unstack()
)


plt.figure(figsize=(12,6))
sns.heatmap(heatmap_data, cmap="Blues", annot=True, fmt=".0f")
plt.title("TSA 2019 Median Traveler Throughput by Day of Week and Month")
plt.show()
```
